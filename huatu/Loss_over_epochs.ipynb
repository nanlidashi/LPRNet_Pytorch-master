```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("out_put.xlsx")
print(df)
```

```python
# 绘制折线图  
plt.figure(figsize=(10, 6))  # 设置图形大小  
plt.plot(df['Epoch'], df['Loss'], color='red', marker='o')  # 绘制折线图，使用圆形标记数据点  
plt.xlabel('Epoch')  # 设置x轴标签  
plt.ylabel('Loss')  # 设置y轴标签  
plt.title('Loss over Epochs')  # 设置标题为 'Accuracy over Epochs'
# 保存图形为 PNG 格式的照片  
plt.savefig('Loss_over_epochs-1.png')  
plt.show()  # 显示图形
```

```python
# 绘制折线图  
plt.figure(figsize=(10, 6))  # 设置图形大小  
plt.plot(df['Epoch'], df['Loss'], color='red', marker='o')  # 绘制折线图，使用圆形标记数据点  
plt.xlabel('Epoch')  # 设置x轴标签  
plt.ylabel('Loss')  # 设置y轴标签  
plt.ylim(0.000, 0.005)  # 设置y轴范围，确保0.001在可视范围内 
plt.title('Loss over Epochs')  # 设置标题为 'Accuracy over Epochs'
plt.axhline(y=0.0007, color='b', linestyle='--')  # 在y轴上绘制96这一条线，使用红色虚线表示 
# 保存图形为 PNG 格式的照片  
plt.savefig('Loss_over_epochs-2.png')
plt.show()  # 显示图形
```
