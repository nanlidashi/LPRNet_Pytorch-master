```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel("out_acc.xlsx")
print(df)
```

```python
# 绘制折线图  
plt.figure(figsize=(10, 6))  # 设置图形大小  
df['Accuracy'] = df['Accuracy'] * 100
plt.plot(df['Epoch'], df['Accuracy'], marker='o')  # 绘制折线图，使用圆形标记数据点  
plt.xlabel('Epoch')  # 设置x轴标签  
plt.ylabel('Accuracy')  # 设置y轴标签  
plt.ylim(0, 100)  # 设置y轴范围，确保96在可视范围内 
plt.title('Accuracy over Epochs')  # 设置标题为 'Accuracy over Epochs'
# 保存图形为 PNG 格式的照片  
plt.savefig('accuracy_over_epochs-1.png')  # 保存文件名为 'accuracy_over_epochs——1.png
plt.show()  # 显示图形
```

```python
# 绘制折线图  
plt.figure(figsize=(10, 6))  # 设置图形大小  
plt.plot(df['Epoch'], df['Accuracy'], marker='o')  # 绘制折线图，使用圆形标记数据点  
plt.xlabel('Epoch')  # 设置x轴标签  
plt.ylabel('Accuracy')  # 设置y轴标签  
plt.ylim(90, 100)  # 设置y轴范围，确保96在可视范围内  
plt.axhline(y=96, color='r', linestyle='--')  # 在y轴上绘制96这一条线，使用红色虚线表示  
plt.title('Accuracy over Epochs')  # 设置标题为 'Accuracy over Epochs'
# 保存图形为 PNG 格式的照片  
plt.savefig('accuracy_over_epochs—2.png')  # 保存文件名为 'accuracy_over_epochs——2.png
plt.show()  # 显示图形
```
